```python
import pandas as pd
import numpy as np
```

```python
features = pd.read_csv("data/elliptic_bitcoin_dataset/elliptic_txs_features.csv", header=None)
edges = pd.read_csv("data/elliptic_bitcoin_dataset/elliptic_txs_edgelist.csv")
classes = pd.read_csv("data/elliptic_bitcoin_dataset/elliptic_txs_classes.csv")
```

```python
classes.columns = ["txId", "class"]
```

```python
df = features.merge(classes, left_on=0, right_on="txId")
```

```python
df['class'] = df['class'].map({'unknown': -1, '1': 1, '2': 0})
df = df[df['class'] != -1]
```

```python
print(df.shape)
df.head()
```

```python
df.to_csv("data/processed_data.csv", index=False)
```
